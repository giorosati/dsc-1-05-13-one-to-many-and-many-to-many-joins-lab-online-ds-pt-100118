## Employees and their Office (a One-to-One join)

Return a list of all of the employees with their first name, last name and the city and state of the office that they work out of (if they have one). Include all employees and order them by their first name, then their last name.

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('data.sqlite', detect_types=sqlite3.PARSE_COLNAMES)
cur = conn.cursor()
```

```python
cur.execute('SELECT firstName, lastName FROM employees ORDER BY firstName, lastName;')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print('Number of results:', len(df))
df.head()
```

```python
#Your code here
#  SELECT ID, FirstName, LastName FROM table GROUP BY(FirstName)
cur.execute('SELECT firstName, lastName, city, state FROM employees JOIN offices GROUP BY employeeNumber ORDER BY firstName, lastName;')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print('Number of results:', len(df))
df.head()
```

## Customers and their Orders (a One-to-Many join)

Return a list of all of the customers. For each customer return a record for each of their order numbers, order dates and statuses.

```python
# Your code here
cur.execute('SELECT customerNumber, customerName, orderNumber, orderDate, status FROM customers JOIN orders USING(customerNumber) ORDER BY customerNumber;')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print('Number of results:', len(df))
df.head(10)
```

## Orders and their Order Details (another One-to-Many join)

Return a list of orders. For each order return a record for each order detail within the order.

```python
# Your code here
cur.execute('SELECT orderNumber, orderLineNumber, productCode, quantityOrdered, priceEach FROM orders JOIN orderDetails USING(orderNumber) ORDER BY orderNumber, orderLineNumber;')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print('Number of results:', len(df))
df.head(10)
```

## Orders, Order details and Product Details (a Many-to-Many Join)

Return a list of the orders with the the order information, and allthe product information for each product in the order. 

```python
# Your code here
cur.execute('SELECT orderNumber, orderDate, requiredDate, shippedDate, status, comments, customerNumber, orderLineNumber, productCode, quantityOrdered, priceEach, productName, productLine, productScale, productVendor, productDescription, quantityInStock, buyPrice, MSRP FROM orders JOIN orderDetails USING(orderNumber) JOIN products USING(productCode) ORDER BY orderNumber, orderLineNumber;')
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print('Number of results:', len(df))
df.head(7)
```
